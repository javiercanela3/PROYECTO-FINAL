# EXTRACCION DE RESULTADOS

Extraemos resultados de la fuente, creando un dataframe unicamente con los valores que nos interesan.
FUENTE: https://www.football-data.co.uk/spainm.php

```python
import pandas as pd
from scipy.stats import poisson
import numpy as np
import seaborn as sns
import warnings
warnings.simplefilter(action='ignore')
pd.options.display.max_columns = None #quitar limite de visualizacion de columnas
```

```python
link = 'https://www.football-data.co.uk/mmz4281/2223/SP1.csv'
pd.read_csv(link).head(3)
```

```python
def temp(años):
    return 'https://www.football-data.co.uk/mmz4281/'+años+'/SP1.csv'
```

```python
def tabla(años):
    #link 04-05 archivo cvs con defefectos
    if años == '0405': tabla = pd.read_excel('temporada0405.xlsx')
    else: tabla = pd.read_csv(temp(años))
    tabla = tabla.dropna(how='all')
    tabla.insert(0, 'Temporada', [años[:2] +'/'+años[2:] for i in range(len(tabla))])
    tabla = tabla[['Temporada', 'Div', 'Date', 'HomeTeam', 'AwayTeam', 'FTHG', 'FTAG','FTR']]
    return tabla
```

```python
años = ['9394', '9495', '9596', '9697', '9798', '9899', '9900',
       '0001', '0102', '0203', '0304', '0405', '0506', '0607',
       '0708', '0809', '0910', '1011', '1112', '1213', '1314',
       '1415', '1516', '1617', '1718', '1819', '1920', '2021',
       '2122', '2223']
```

```python
resultados = tabla(años[0])
for año in años[1:]:
    resultados = pd.concat([tabla(año),resultados])

resultados.FTHG = resultados.FTHG.astype('int64')
resultados.FTAG = resultados.FTAG.astype('int64')
resultados.Date = pd.to_datetime(resultados.Date, dayfirst=True, infer_datetime_format=True)   
resultados = resultados.sort_values(by='Date', ascending = False)
resultados
```

```python
sorted(list(resultados.HomeTeam.unique()))
```

```python
resultados.HomeTeam.replace({'Villareal':'Villarreal', 'Vallecano':'Rayo Vallecano'}, inplace=True)
resultados.AwayTeam.replace({'Villareal':'Villarreal', 'Vallecano':'Rayo Vallecano'}, inplace=True)
resultados.Div.replace({'SP1': 'La Liga'}, inplace=True)
```

```python
from datetime import datetime
from datetime import timedelta
def fecha_hace_n_años(n=3): return (datetime.now() - timedelta(days=365*n))
fecha_hace_n_años()
```

```python
def historico_2_equipos(team_A, team_B, ultimos_n_años='all'):
    historico = resultados[((resultados.HomeTeam == team_A) & (resultados.AwayTeam == team_B)) | 
          ((resultados.HomeTeam == team_B) & (resultados.AwayTeam == team_A))]
    if ultimos_n_años != 'all':
        historico = historico[historico.Date > fecha_hace_n_años(ultimos_n_años)]
    
    team_A_local = historico[historico.HomeTeam==team_A]
    team_B_local = historico[historico.HomeTeam==team_B]
    
    df = pd.DataFrame()
    df['Equipo'] = [team_A, team_B]
    df['VL'] = [list(team_A_local.FTR.values).count('H'), list(team_B_local.FTR.values).count('H')]
    df['EL'] = [list(team_A_local.FTR.values).count('D'), list(team_B_local.FTR.values).count('D')]
    df['DL'] = [list(team_A_local.FTR.values).count('A'), list(team_B_local.FTR.values).count('A')]
    df['VV'] = [list(team_B_local.FTR.values).count('A'), list(team_A_local.FTR.values).count('A')]
    df['EV'] = [list(team_B_local.FTR.values).count('D'), list(team_A_local.FTR.values).count('D')]
    df['DV'] = [list(team_B_local.FTR.values).count('H'), list(team_A_local.FTR.values).count('H')]
    
    teams = [team_A, team_B]
    
    for i, e in enumerate(teams):
        print('------------------\n----',e, '---\n------------------')
        print(len(historico), 'enfrentamientos desde', historico.Date.iloc[-1])
        print('------------------')

        print(df.VL[i], 'victorias local')
        print(df.EL[i], 'empates local')
        print(df.DL[i], 'derrotas local')
        print(df.VV[i], 'victorias visitante')
        print(df.EV[i], 'empates visitante')
        print(df.DV[i], 'derrotas visitante') 
    return historico, df
historico, estadisticas = historico_2_equipos('Real Madrid', 'Valencia', 3)
estadisticas
```

```python
def historico_1_equipo(team, ultimos_n_enfrentamientos='all'):
    historico = resultados[(resultados.HomeTeam == team) | 
          (resultados.AwayTeam == team)]
    if ultimos_n_enfrentamientos != 'all':
        historico = historico[:ultimos_n_enfrentamientos]
    
    team_local = historico[historico.HomeTeam==team]
    team_visit = historico[historico.AwayTeam==team]
    
    
    df = pd.DataFrame()
    df['Equipo'] = [team]
    df['VL'] = [list(team_local.FTR.values).count('H')]
    df['DL'] = [list(team_local.FTR.values).count('A')]
    df['EL'] = [list(team_local.FTR.values).count('D')]
    df['VV'] = [list(team_visit.FTR.values).count('A')]
    df['DV'] = [list(team_visit.FTR.values).count('H')]
    df['EV'] = [list(team_visit.FTR.values).count('D')]
    
    print('------------------\n----',team, '---\n------------------')
    print(len(historico), 'enfrentamientos desde', historico.Date.iloc[-1])
    print('------------------')
    print(df.VL[0], 'victorias local')
    print(df.EL[0], 'empates local')
    print(df.DL[0], 'derrotas local')
    print(df.VV[0], 'victorias visitante')
    print(df.EV[0], 'empates visitante')
    print(df.DV[0], 'derrotas visitante\n')
    
    return historico, df, team_local, team_visit
historico, estadisticas, team_local, team_visit = historico_1_equipo('Real Madrid', 38)
estadisticas
```
